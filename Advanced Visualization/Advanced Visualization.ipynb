## Stacked Histogram

```python
plt.hist(df.BudgetMillions)
plt.show()
```

```python
plt.hist(df[df.Genre=='Drama'].BudgetMillions,bins=15)
plt.hist(df[df.Genre=='Action'].BudgetMillions,color='green',bins=15)
plt.hist(df[df.Genre=='Thriller'].BudgetMillions,color='red',bins=15);
```

```python
plt.hist([df[df.Genre=='Drama'].BudgetMillions,df[df.Genre=='Action'].BudgetMillions,df[df.Genre=='Thriller'].BudgetMillions],bins=10)
plt.show()
```

```python
plt.hist([df[df.Genre=='Drama'].BudgetMillions,\
          df[df.Genre=='Action'].BudgetMillions,\
          df[df.Genre=='Thriller'].BudgetMillions],bins=10,stacked=True)
plt.show()
```

```python
list1=list()
mylabel=[]
for i in df.Genre.cat.categories:
    list1.append(df[df.Genre==i].BudgetMillions)
    mylabel.append(i)
```

```python
plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabel);
plt.legend()
```
