## Creating a Facet grid

```python
f=sns.FacetGrid(df,row='Genre',col='Year',hue='Genre')
f=f.map(plt.scatter,'CriticRating','AudienceRating')
```

```python
# Can populate with any type of plot
f=sns.FacetGrid(df,row='Genre',col='Year',hue='Genre')
f=f.map(plt.hist,'BudgetMillions')
```

```python
f=sns.FacetGrid(df,row='Genre',col='Year',hue='Genre')
kws=dict(s=50,linewidth=0.5,edgecolor='black')
f=f.map(plt.scatter,'CriticRating','AudienceRating',**kws)
```

## Coordinates and Diagonal

```python
## Controlling axes and Adding Diagonals
f=sns.FacetGrid(df,row='Genre',col='Year',hue='Genre')
kws=dict(s=50,linewidth=0.5,edgecolor='black')
f=f.map(plt.scatter,'CriticRating','AudienceRating',**kws)
f.set(xlim=(0,100),ylim=(0,100))
for ax in f.axes.flat:
    ax.plot((0,100),(0,100),c='gray',ls='--')
f.add_legend
```
