```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize']=8,4
```

```python
df=pd.read_csv('car.csv')
df
```

```python
price=[4000,5000,7000,22000,3500,4500,7500,7000,6250,9700]
```

```python
df['price']=price
df
```

```python
df.drop('Price',axis=1,inplace=True)
df
```

```python
# Distribution:
vis1=sns.distplot(df['Doors'])
```

```python
visu2=sns.distplot(df['Odometer (KM)'])
```

```python
visu2=sns.distplot(df['Odometer (KM)'],bins=10)
```

```python
# Boxplot
visu3=sns.boxplot(data=df,x='Doors',y='price')
```

```python
visu4=sns.boxplot(data=df,x='Doors',y='Odometer (KM)')
```

```python
visu5=sns.lmplot(data=df,x='Doors',y='Odometer (KM)')
```

```python
visu6=sns.lmplot(x='Odometer (KM)',y='Doors',data=df,hue='Make')
```

```python
v7=sns.lmplot(x='Doors',y='price',data=df,hue='Make')
```
